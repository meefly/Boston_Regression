```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
dataBoston =  load_boston()
df  =  pd.DataFrame(
    dataBoston['data'],
    columns = dataBoston['feature_names']
)
df['price'] = dataBoston['target']
df
```

```python
korelasi = df.corr()
plt.imshow(korelasi, cmap='BuPu_r')
plt.colorbar()
plt.xticks(np.arange(14),list(df.columns),rotation = 45)
plt.yticks(np.arange(14),list(df.columns))
plt.show()
```

ML linear regression

```python
model = linear_model.LinearRegression()
model.fit(df[['ZN','CHAS','RM','DIS','B']], df['price'])
# model.predict(df[['ZN','CHAS','RM','DIS','B']])
model.predict(df.head(1)[['ZN','CHAS','RM','DIS','B']])
# df.head(1)[['ZN','CHAS','RM','DIS','B']]
```
